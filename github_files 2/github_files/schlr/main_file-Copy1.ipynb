```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import re

import fasttext
import numpy as np

# Download the NLTK data for POS tagging
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

class extract_pos:

    def __init__(self):
        self.pattern = r'[^a-zA-Z0-9\s]'

    def remove_word_with_only_spl_char(self, word_list):
        pattern = r'[^a-zA-Z0-9\s]'
        new = []
        for i in word_list:
            result = re.sub(pattern, '', i)
            if result != '':
                new.append(i)
        return new
    
    def extract_adjectives(self,text):
        # Tokenize the text into words
        words = word_tokenize(text)
    
        # Perform POS tagging
        pos_tags = pos_tag(words)
    
        # Extract adjectives based on POS tagging
        adjectives = [word for word, pos in pos_tags if pos.startswith('JJ')]
        
    
        # Use re.sub to replace special characters with an empty string
        adjectives = self.remove_word_with_only_spl_char(adjectives)
        return adjectives
    
    def extract_verbs(self,text):
        # Tokenize the text into words
        words = word_tokenize(text)
    
        # Perform POS tagging
        pos_tags = pos_tag(words)
    
        # Extract verbs based on POS tagging
        verbs = [word for word, pos in pos_tags if pos.startswith('VB')]

        verbs = self.remove_word_with_only_spl_char(verbs)
    
        return verbs

    def extract_needed_pos(self, text):
        what_pos = 'JJ'
        pos_return = extract_pos_text.extract_adjectives(text)
        if len(pos_return)==0:
            pos_return = extract_pos_text.extract_verbs(text)
            what_pos = 'VB'
        return (pos_return, what_pos)        
        

class FastTextRelevanceFinder:
    def __init__(self, model_path):
        # Load pre-trained FastText model
        self.model = fasttext.load_model(model_path)

    def get_word_vector(self, word):
        """
        Get the word vector using the FastText model.
        """
        try:
            return self.model[word]
        except KeyError:
            # Handle the case when the word is not in the vocabulary
            return np.zeros(self.model.get_dimension())

    def find_relevant_words(self, paragraph, target_word, k=5):
        """
        Find relevant words in a paragraph for a target word using FastText.
        """
        # Tokenize the paragraph into words
        tokens = paragraph.split()

        # Get the word vector for the target word
        target_vector = self.get_word_vector(target_word)

        # Find relevant words based on similarity ( its not found)
        relevant_words = []
        for token in tokens:
            vector = self.get_word_vector(token)
            similarity = np.dot(target_vector, vector) / (np.linalg.norm(target_vector) * np.linalg.norm(vector))
            relevant_words.append((token, similarity))

        # Sort the words by similarity and get the top k
        relevant_words = sorted(relevant_words, key=lambda x: x[1], reverse=True)[:k]

        return relevant_words

# Example text
example_text = "I am currently enjoying this album 'Listen to Eason Chan.'"
extract_pos_text = extract_pos()
# Extract adjectives from the example text
extracted_needed_pos = extract_pos_text.extract_needed_pos(example_text)
#verb = extract_pos_text.extract_verbs(example_text)
print(extracted_needed_pos)


```
